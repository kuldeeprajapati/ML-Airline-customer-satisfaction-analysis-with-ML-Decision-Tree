```python
#Create a function to display precision score, recall score, accuracy, classification report, confusion matrix, F1 Score
```

```python
from sklearn.metrics import recall_score,f1_score,accuracy_score,confusion_matrix,classification_report,precision_score
```

```python
def run_model(model,x_train,y_train,x_test,y_test):
    model.fit(x_train,y_train.ravel())
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("pricison_score: ",precision_score(y_test, y_pred))
    print("recall_score: ",recall_score(y_test, y_pred))
    print("Accuracy = {}".format(accuracy))
    print(classification_report(y_test,y_pred,digits=5))
    print(confusion_matrix(y_test,y_pred))
```

```python
run_model(model,x_train, y_train, x_test, y_test)
```
